```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/gourisharma32/hr-employee-attrition/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
print(df.shape)
print(df.columns)
df.head()
```

```python
# Missing values
df.isnull().sum()

# Duplicate rows
df.duplicated().sum()

# Data types
df.info()
```

```python
df.nunique().sort_values()
```

```python
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1, inplace=True)

df.shape
```

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

df[['Attrition', 'Attrition_Flag']].head()
```

```python
df['Attrition_Flag'].value_counts()
```

```python
attrition_rate = df['Attrition_Flag'].mean() * 100
print(round(attrition_rate, 2))
```

```python
df['MonthlyIncome'].mean()
```

```python
df['YearsAtCompany'].mean()
```

```python
tenure_attrition = df.groupby('YearsAtCompany')['Attrition_Flag'].mean() * 100
print(tenure_attrition.head(10).round(2)) #years at company
```

```python
pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100  #percent of people leaving in each group
```

```python
corr = df.corr(numeric_only=True)

corr['Attrition_Flag'].sort_values(ascending=False)
```

```python
df['OverTime_Flag'] = df['OverTime'].map({'Yes':1, 'No':0})
```

```python
df.drop('MonthlyRate', axis=1, inplace=True)
```

```python
# Age groups
df['AgeGroup'] = pd.cut(df['Age'], bins=[18,25,35,45,60],
                        labels=['18-25','26-35','36-45','46-60'])

# Income groups
df['IncomeGroup'] = pd.qcut(df['MonthlyIncome'], 4,
                           labels=['Low','Medium','High','Very High'])
```

```python
df.to_csv('/kaggle/working/hr_clean.csv', index=False)
```

```python
dept_attrition = df.groupby('Department')['Attrition_Flag'].mean().sort_values(ascending=False) * 100
dept_attrition = dept_attrition.round(2)

print(dept_attrition)
```
